```python
import requests
import pandas as pd
from io import StringIO
import datetime
import sqlite3

def Update_price_daily(date):
    date_I_want = date.strftime('%Y%m%d')
    r = requests.post('http://www.twse.com.tw/exchangeReport/MI_INDEX?response=csv&date=' + date_I_want  + '&type=ALLBUT0999')
    content = r.text.replace('=', '')
    lines = content.split('\n')
    lines = list(filter(lambda l:len(l.split('",')) > 10, lines))
    content = "\n".join(lines)
    if content == '':
        return None
    df = pd.read_csv(StringIO(content))
    df = df.astype(str)
    df['date'] = pd.to_datetime(date)
    df = df.set_index(['證券代號', 'date','證券名稱'])
    df =df.iloc[:,0:9]
    df = df[df.columns[df.isnull().all() == False]]
    df.to_csv('Update_price_daily.csv', encoding='utf_8_sig')
    df = pd.read_csv('Update_price_daily.csv', index_col=['證券代號'])
    conn = sqlite3.connect('Update_price_daily.sqlite3')
    df.to_sql('Update_price_daily', conn, if_exists='replace')
    return df
Update_price_daily(datetime.datetime(2019,12,4))  
```
